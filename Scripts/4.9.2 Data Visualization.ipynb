# Creating and exporting bar charts

```python
# Create bar chart for orders_day_of_the_week
bar = df_ordprodcust['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['navy','blue','royalblue','deepskyblue', 'turquoise', 'darkcyan', 'seagreen']) 
```

```python
# Create bar chart for orders per customer loyalty group
bar_loyalty = df_ordprodcust['loyalty_flag'].value_counts().plot.bar(color = ['navy','blue','royalblue'])
```

```python
# Export created bar chart
bar.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Export loyalty bar chart
bar_loyalty.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'bar_loyalty.png'))
```

# Creating and exporting histograms & scatterplots

```python
# Create histogram for prices with 25 bins
hist_prices1 = df_ordprodcust['prices'].plot.hist(bins = 25, color = 'navy')
```

```python
# Create histogram for prices with 70 bins
hist_prices2 = df_ordprodcust['prices'].plot.hist(bins = 70, color = 'navy')
```

```python
# Create histogram for ordertime_hour_of_day
hist_hod = df_ordprodcust['ordertime_hour_of_day'].plot.hist(bins = 24, color = 'navy')
```

The histogram for the ordertime_hour_of_day shows that most orders occur between around 09:00 and 17:00.

```python
# Create scatterplot of prices
scatt_prices = sns.scatterplot(x = 'prices', y = 'prices', data = df_ordprodcust, color = 'navy')
```

```python
# Create scatterplot for age and spending power
scatt_ageincome = sns.scatterplot(x = 'age', y = 'income', data = df_ordprodcust, color = 'navy')
```

The scatterplot shows that there seems to be higher spending power for customers aged 40 and above, as there are various observations above the 400000 income mark and there are no observations of this income in the ages below.

# Create and export line charts

```python
# Creating sample for line chart step 1
np.random.seed(4)
```

```python
# Creating sample for line chart step 2
dev = np.random.rand(len(df_ordprodcust)) <= 0.7
```

```python
# Creating big sample
big = df_ordprodcust[dev]
```

```python
# Creating small smaple
small = df_ordprodcust[~dev]
```

```python
# Checking whole df length
len(df_ordprodcust)
```

```python
# Checking length of both samples to see if they match up
len(big) + len(small)
```

```python
# Creating small df from small sample with relevant columns for line chart dow/prices
df_smallsample_dowp = small[['orders_day_of_week','prices']]
```

```python
# Creating line chart for day of week/prices
line = sns.lineplot(data = df_smallsample_dowp, x = 'orders_day_of_week',y = 'prices', color = 'navy')
```

```python
# Creating small df from small sample with relevant columns for line chart dow/prices
df_smallsample_hodp = small[['ordertime_hour_of_day','prices']]
```

```python
# Creating line chart for hour of day/prices
line2 = sns.lineplot(data = df_smallsample_hodp, x = 'ordertime_hour_of_day',y = 'prices', color = 'navy')
```

```python
# Creating small df from small sample with relevant columns for line chart age/number of dependants
df_smallsample_aged = small[['age','n_dependants']]
```

```python
# Creating line chart for hour of day/prices
line3 = sns.lineplot(data = df_smallsample_aged, x = 'age',y = 'n_dependants', color = 'navy')
```

There does not seem to be a relationship between the age of customers and the number of dependants they have. 

```python
# Export line chart dow/prices
line.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'line_dowp'))
```

```python
# Export line chart hod/prices
line2.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'line_hodp'))
```

```python
# Export line chart age/dependants
line3.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'line_aged'))
```
